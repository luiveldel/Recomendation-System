## 3. EDA

### Recommending based on counts

```python
# How many ratings for each place?
rating_count = pd.DataFrame(ratings.groupby('placeID')['rating'].count())
rating_count.sort_values(by='rating', ascending=False).head(5)
```

```python
# The most reviewed is placeID 135085
# Top 5 most rated places
top5_rated_places = pd.DataFrame(rating_count.sort_values(by='rating', ascending=False).head(5))
```

```python
# Any similarity between the top places on their cuisine?
top5_cuisines = pd.merge(top5_rated_places, cuisine, how='inner', on='placeID')
top5_cuisines
```
